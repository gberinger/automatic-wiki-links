```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
results_df = pd.read_csv('results.csv')
results_shuffle_df = pd.read_csv('results_shuffle.csv')
plt.plot(results_df['epoch'], results_df['top1_acc'], 'bo-', label='No shuffle')
plt.plot(results_shuffle_df['epoch'], results_shuffle_df['top1_acc'], 'go-', label='Shuffle')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Top-1 accuracy (context=4, alpha=0.15, beta=0.03)')
plt.legend()
plt.savefig('acc.png')
plt.show()
```
