```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
contexts = range(1, 11)
for c in contexts:
    max_acc = 0
    for v in ["en", "en_core_web_sm", "en_core_web_md", "en_core_web_lg", "en_vectors_web_lg"]:
        # without shuffle
        results_df = pd.read_csv('results_{}_no_shuffle_{}.csv'.format(v, c))
        plt.plot(results_df['epoch'], results_df['top1_acc'], 'o-', label=v)
        max_acc = max(max_acc,results_df['top1_acc'].max()) 
        
        # with shuffle
        results_df = pd.read_csv('results_{}_shuffle_{}.csv'.format(v, c))
        plt.plot(results_df['epoch'], results_df['top1_acc'], 'o--', label=v+" shuffle")
        max_acc = max(max_acc,results_df['top1_acc'].max()) 

    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Top-1 accuracy, context={}, alpha=0.15'.format(c))
    plt.legend(bbox_to_anchor=(1.1, 1.05)) 
    plt.savefig('top1_acc_all_vocabularies_{}.png'.format(c))
    plt.show()
    print('Context={}, max_acc={}'.format(c, max_acc))
```

```python
contexts = range(1, 11)
for c in contexts:
    max_acc = 0
    for v in ["en_core_web_md", "en_core_web_lg", "en_vectors_web_lg" ]:
        # without shuffle
        results_df = pd.read_csv('results_{}_no_shuffle_{}.csv'.format(v, c))
        plt.plot(results_df['epoch'], results_df['top1_acc'], 'o-', label=v)
        max_acc = max(max_acc,results_df['top1_acc'].max()) 
        
        # with shuffle
        results_df = pd.read_csv('results_{}_shuffle_{}.csv'.format(v, c))
        plt.plot(results_df['epoch'], results_df['top1_acc'], 'o--', label=v+" shuffle")
        max_acc = max(max_acc,results_df['top1_acc'].max()) 

    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Top-1 accuracy, context={}, alpha=0.15'.format(c))
    plt.legend(bbox_to_anchor=(1.1, 1.05)) 
    plt.savefig('top1_acc_best_vocabularies_{}.png'.format(c))
    plt.show()
    print('Context={}, max_acc={}'.format(c, max_acc))
```

```python
contexts = range(1, 11)
for c in contexts:
    for v in ["en_core_web_md", "en_core_web_lg", "en_vectors_web_lg" ]:
        max_acc = 0
        for p in ["huge", "huge_beta"]:
            for s in ["shuffle", "no_shuffle"]:
                results_df = pd.read_csv('results_{}_{}_{}_{}.csv'.format(p, v, s, c))
                label = v
                style = 'o-'
                if p == "huge_beta":
                    label += ", beta=0.03"
                if s == "shuffle":
                    label += ", shuffle"
                    style = 'o--'
                label += "(max = {:.3f})".format(results_df['top1_acc'].max())
                
                plt.plot(results_df['epoch'], results_df['top1_acc'], style, label=label)
                max_acc = max(max_acc,results_df['top1_acc'].max()) 
           

        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.title('Top-1 accuracy, context={}, model={}, alpha=0.15'.format(c,v))
        plt.legend(bbox_to_anchor=(1.1, 1.05)) 
        plt.savefig('top1_acc_with_beta_{}_{}.png'.format(c,v))
        plt.show()
        print('Context={}, max_acc={:.4f}'.format(c, max_acc))
```
