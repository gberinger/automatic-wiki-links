```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
results_df = pd.read_csv('results.csv')
no_train_df = pd.read_csv('results_no_train.csv')
```

```python
plt.scatter(no_train_df['context'], no_train_df['top1_acc'], label='Before training')
plt.scatter(results_df['context'], results_df['top1_acc'], label='After 1 epoch of training')
plt.xlabel('Context')
plt.ylabel('Accuracy')
plt.title('Top-1 accuracy based on context (epochs=1)')
plt.legend()
plt.savefig('plot.png')
plt.show()
```
