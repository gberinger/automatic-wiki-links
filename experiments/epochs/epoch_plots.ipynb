```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
results_df = pd.read_csv('results_3.csv')
```

```python
plt.scatter(results_df['epoch'], results_df['cos_dist'])
plt.xlabel('Epoch')
plt.ylabel('Avg cos dist')
plt.title('Avg cos dist, context (context=3)')
plt.savefig('plot_3.png')
plt.show()
```

```python
contexts = range(1, 11)
for c in contexts:
    results_df = pd.read_csv('results_{}.csv'.format(c))
    plt.scatter(results_df['epoch'], results_df['top1_acc'])
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Top-1 accuracy, context={}, alpha=0.15'.format(c))
    plt.savefig('top1_acc_{}.png'.format(c))
    plt.show()
    print('Context={}, max_acc={}'.format(c, results_df['top1_acc'].max()))
```

```python
results_df = pd.read_csv('results_4_cos_dist.csv')
plt.scatter(results_df['epoch'], results_df['top1_acc'])
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Top-1 accuracy, context=4, alpha=cos_dist')
plt.savefig('top1_acc_4_cos_dist.png')
plt.show()
print('Context=4, max_acc={}'.format(results_df['top1_acc'].max()))
```
