```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
results1_df = pd.read_csv('results_1.csv')
results4_df = pd.read_csv('results_4.csv')
plt.scatter(results1_df['beta'], results1_df['top1_acc'], label='Epoch 1')
plt.scatter(results4_df['beta'], results4_df['top1_acc'], label='Epoch 4')
plt.xlabel('Beta')
plt.ylabel('Accuracy')
plt.title('Top-1 accuracy (context=4, alpha=0.15)')
plt.legend()
plt.savefig('acc.png')
plt.show()
```
