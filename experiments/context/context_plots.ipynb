```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
no_train_df = pd.read_csv('results_no_train.csv')
```

```python
alpha_015_df = pd.read_csv('results_alpha_015.csv')
alpha_cos_dist_df = pd.read_csv('results_alpha_cos_dist.csv')
plt.scatter(no_train_df['context'], no_train_df['top1_acc'], label='Before training')
plt.scatter(alpha_015_df['context'], alpha_015_df['top1_acc'], label='Alpha: 0.15')
plt.scatter(alpha_cos_dist_df['context'], alpha_cos_dist_df['top1_acc'], label='Alpha: cosinus distance')
plt.xlabel('Context')
plt.ylabel('Accuracy')
plt.title('Top-1 accuracy (epochs=1, alpha=0.15)')
plt.legend()
plt.savefig('acc.png')
plt.show()
```

```python
results_df = pd.read_csv('results_alpha_cos_dist.csv')
plt.scatter(no_train_df['context'], no_train_df['top1_acc'], label='Before training')
plt.scatter(results_df['context'], results_df['top1_acc'], label='After 1 epoch of training')
plt.xlabel('Context')
plt.ylabel('Accuracy')
plt.title('Top-1 accuracy (epochs=1, alpha=cos_dist)')
plt.legend()
plt.savefig('acc_alpha_cos_dist.png')
plt.show()
```
