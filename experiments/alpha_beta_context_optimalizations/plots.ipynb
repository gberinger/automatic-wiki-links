```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import json
```

### Motivation for the experiment

Use default `alpha`, `beta` and `context` values and separately test other values for each parameter from given ranges:
* for `alpha`: <0;0.5>
* for `beta`: <0;0.1>
* for `context`: <1;10>

After each iteration take best parameters and use them next iteration. Do 5 iterations.

```python
experiments = ["alpha", "beta", "context"]
for opt_number in range(5):
    print("\nOpt {}\n".format(opt_number))
    for experiment in experiments:
        max_acc = 0
        experiment_path = "../{}_opt_{}_larger_no_shuffle/".format(experiment, opt_number)
        results_df = pd.read_csv(experiment_path + 'results.csv')
        plt.plot(results_df[experiment], results_df['top1_acc'], 'o-')
        max_acc = max(max_acc,results_df['top1_acc'].max()) 
        plt.xlabel(experiment)
        plt.ylabel('Accuracy')
        plt.title('Top-1 accuracy') 
        plt.savefig('top1_acc_{}_{}.png'.format(experiment, opt_number))
        plt.show()
        with open(experiment_path + 'args.txt') as f:
            data = json.load(f)
            for key, value in data.items():
                if key in experiments and key != experiment:
                  print(key, value)
        print('max_acc={} at {} = {} '.format(max_acc, experiment,results_df[experiment][results_df["top1_acc"].idxmax()]))
    
```

### Interpretation of results

Create plot with best accuracy for each tested parameter and check progress trend.

```python
experiments = ["alpha", "beta", "context"]
maxes = {}

for experiment in experiments:
    maxes[experiment]=[]
    for opt_number in range(0,7):
        max_acc = 0
        experiment_path = "../{}_opt_{}_larger_no_shuffle/".format(experiment, opt_number)
        results_df = pd.read_csv(experiment_path + 'results.csv')
        max_acc = max(max_acc,results_df['top1_acc'].max())
        maxes[experiment].append(max_acc)
    
    plt.plot(maxes[experiment])
    plt.xlabel("Opt " + experiment)
    plt.ylabel('Accuracy')
    plt.title('Top-1 accuracy') 
    plt.savefig('opt_acc_{}.png'.format(experiment))
    plt.show()
    
```

### Next experiments

Third optimization was the best, use it for next experiments.

Check average result from training with shuffle mode enabled, run 30 experiments for each parameter.

```python
experiments = ["alpha", "beta", "context"]
for opt_number in range(3,4):
    print("\nOpt {}\n".format(opt_number))
    for experiment in experiments:
        max_acc_normal = 0
        experiment_path = "../{}_opt_{}_larger_no_shuffle/".format(experiment, opt_number)
        results_df = pd.read_csv(experiment_path + 'results.csv')
        plt.plot(results_df[experiment], results_df['top1_acc'], 'o-')
        max_acc_normal = max(max_acc_normal,results_df['top1_acc'].max())
        results_df_normal = results_df
        top1_acc = 0 * results_df['top1_acc']
        offset = 1
        num_shuffles = 30
        for i in range(offset, num_shuffles + offset):
            experiment_path = "../{}_opt_{}_larger_shuffle_{}/".format(experiment, opt_number,i)
            results_df = pd.read_csv(experiment_path + 'results.csv')
            top1_acc = top1_acc + results_df['top1_acc']
        
        top1_acc = top1_acc/num_shuffles
        
        plt.plot(results_df[experiment], top1_acc, 'o-', label="shuffled avg of {} runs".format(num_shuffles))
        plt.xlabel(experiment)
        plt.ylabel('Accuracy')
        plt.title('Top-1 accuracy') 
        plt.savefig('top1_acc_mixed_{}_{}.png'.format(experiment, opt_number))
        plt.legend()
        plt.show()
        with open(experiment_path + 'args.txt') as f:
            data = json.load(f)
            for key, value in data.items():
                if key in experiments and key != experiment:
                  print(key, value)
        print('Without shuffle max_acc={} at {} = {} '.format(max_acc_normal, experiment,results_df_normal[experiment][results_df_normal["top1_acc"].idxmax()]))
        print('With shuffle max_acc={} at {} = {} '.format(top1_acc.max(), experiment,results_df_normal[experiment][top1_acc.idxmax()]))
    
```

### Interpretation of results

Shuffle allows for using higher `beta` values. It was expected as `beta` parameters is something like negative reward for bad prediction. During training model is overfitted for each keyword one by one. It causes degradation of previously acquired knowledge.

### Next steps

When we have semi-optimized parameters we can check if many epochs of training improves model.

Run same experiments for different number of epochs, from range <1;10>. Compare cumulated results from each experiment.

```python
tested_epochs = [1,3,5,7]
opt_numbers = [3]
for opt_number in opt_numbers:
    print("\nOpt {}\n".format(opt_number))
    for experiment in experiments:
        max_acc_normal = 0
        for epoch in tested_epochs:
            experiment_path_prefix = "../{}_opt_{}_larger_many_epochs_{}".format(experiment, opt_number, epoch)
            experiment_path = experiment_path_prefix + "_no_shuffle/"
            results_df = pd.read_csv(experiment_path + 'results.csv')
            #plt.plot(results_df[experiment], results_df['top1_acc'], 'o-', label="epoch_{} no shuffle".format(epoch))
            max_acc_normal = max(max_acc_normal,results_df['top1_acc'].max())
            results_df_normal = results_df
            top1_acc = 0 * results_df['top1_acc']
            offset = 1
            num_shuffles = 30
            for i in range(offset, num_shuffles + offset):
                experiment_path = experiment_path_prefix + "_shuffle_{}/".format(i)
                results_df = pd.read_csv(experiment_path + 'results.csv')
                top1_acc = top1_acc + results_df['top1_acc']
        
            top1_acc = top1_acc/num_shuffles
        
            plt.plot(results_df[experiment], top1_acc, 'o-', label="epoch_{} shuffled avg of {} runs".format(epoch, num_shuffles))

        plt.xlabel(experiment)
        plt.ylabel('Accuracy')
        plt.title('Top-1 accuracy') 
        plt.savefig('top1_acc_many_contexts_{}_{}.png'.format(experiment, opt_number))
        plt.legend()
        plt.show()
        with open(experiment_path + 'args.txt') as f:
            data = json.load(f)
            for key, value in data.items():
                if key in experiments and key != experiment:
                  print(key, value)
        print('Without shuffle max_acc={} at {} = {} '.format(max_acc_normal, experiment,results_df_normal[experiment][results_df_normal["top1_acc"].idxmax()]))
        print('With shuffle max_acc={} at {} = {} '.format(top1_acc.max(), experiment,results_df_normal[experiment][top1_acc.idxmax()]))
    
```
